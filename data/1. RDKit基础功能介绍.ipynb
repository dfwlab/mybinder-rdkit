```python
from rdkit import Chem
from rdkit.Chem import Draw
```

# 1. 读取SMILES¶

```python
mol = Chem.MolFromSmiles("CC(=O)Oc1ccccc1C(=O)O")
Draw.MolToImage(mol)
```

# 2. 读取Mol文件

```python
mol = Chem.MolFromMolFile('data/example.mol')
Draw.MolToImage(mol)
```

# 3. 读取SDF文件

```python
suppl = Chem.SDMolSupplier('data/example.sdf')
# 遍历分子并提取注释信息
for mol in suppl:
    if mol is not None:
        print("Molecular CID:", mol.GetProp("PUBCHEM_COMPOUND_CID"))
        print("Molecular IUPAC:", mol.GetProp("PUBCHEM_IUPAC_OPENEYE_NAME"))
Draw.MolToImage(mol)
```

# 4. 分子信息

```python
from rdkit import Chem
from rdkit.Chem import Descriptors

mol = Chem.MolFromSmiles("CC(=O)Oc1ccccc1C(=O)O")

# 分子基本信息和描述
atom_count = mol.GetNumAtoms()  # 原子数
bond_count = mol.GetNumBonds()  # 键数
mol_weight = Descriptors.MolWt(mol)  # 分子量
logP = Descriptors.MolLogP(mol)  # LogP
hba = Descriptors.NumHAcceptors(mol)  # 氢键受体 (HBA)
hbd = Descriptors.NumHDonors(mol)  # 氢键供体 (HBD)

print(f"Number of Atoms: {atom_count}")
print(f"Number of Bonds: {bond_count}")
print(f"Molecular Weight: {mol_weight:.2f}")
print(f"LogP: {logP:.2f}")
print(f"HBA: {hba}")
print(f"HBD: {hbd}")
```

# 5. 输出

```python
from rdkit import Chem
mol = Chem.MolFromSmiles("CC(=O)Oc1ccccc1C(=O)O")

# 输出 SMILES
smiles = Chem.MolToSmiles(mol)
print("SMILES:", smiles)
```

```python
from rdkit import Chem
mol = Chem.MolFromSmiles("CC(=O)Oc1ccccc1C(=O)O")

# 输出 mol block
mol_block = Chem.MolToMolBlock(mol)
print("Mol Block:\n", mol_block)
```
